#### Looping through list

```python
thislist= ['apple','banana','mango']
```

#### For loop

```python
for x in thislist:
    print(x)
```

```python
for i in range(len(thislist)):
    print(thislist[i])
```

#### While loop

```python
thislist= ['apple','banana','mango']
i = 0
while i < len(thislist):
    print(thislist[i])
    i = i + 1
```

#### List comprehension

```python
thislist= ['apple','banana','mango']
[print(x) for x in thislist]
```

```python
fruits= ['apple', 'banana', 'cherry', 'kiwi', 'mango']
newlist= []
for x in fruits:
    if 'a' in x:
        newlist.append(x)

print(newlist)     
```

#### Defining function

```python
def myfunc(n):
    return abs(n - 50)
thislist= [100, 50, 65, 82, 23]
thislist.sort(key= myfunc)
print(thislist)
```
