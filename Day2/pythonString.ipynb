# Python Strings Tutorial

This notebook covers string operations, formatting, and methods in Python.

## Basic String Creation

Strings in Python are surrounded by either single or double quotation marks.

```python
name = 'Brad'
age = 37

print(f"Name: {name}")
print(f"Age: {age}")
```

## String Concatenation

You can combine strings using the + operator.

```python
# Concatenate
print('Hello, my name is ' + name + ' and I am ' + str(age))
```

## String Formatting

Python offers several ways to format strings.

```python
# Arguments by position
print('My name is {name} and I am {age}'.format(name=name, age=age))
```

```python
# F-Strings (3.6+) - Most modern and preferred method
print(f'Hello, my name is {name} and I am {age}')
```

## String Methods

Python provides many built-in methods to work with strings.

```python
s = 'helloworld'
print(f"Original string: {s}")
```

### Case Manipulation

```python
# Capitalize string
print(f"Capitalize: {s.capitalize()}")

# Make all uppercase
print(f"Uppercase: {s.upper()}")

# Make all lower
print(f"Lowercase: {s.lower()}")

# Swap case
print(f"Swapcase: {s.swapcase()}")
```

### String Analysis

```python
# Get length
print(f"Length: {len(s)}")

# Count occurrences
sub = 'l'
print(f"Count of '{sub}': {s.count(sub)}")

# Find position
print(f"Position of 'r': {s.find('r')}")
```

### String Testing

```python
# Starts with
print(f"Starts with 'hello': {s.startswith('hello')}")

# Ends with
print(f"Ends with 'd': {s.endswith('d')}")

# Is all alphanumeric
print(f"Is alphanumeric: {s.isalnum()}")

# Is all alphabetic
print(f"Is alphabetic: {s.isalpha()}")

# Is all numeric
print(f"Is numeric: {s.isnumeric()}")
```

### String Manipulation

```python
# Replace
print(f"Replace 'world' with 'everyone': {s.replace('world', 'everyone')}")

# Split into a list
sentence = "hello world python"
print(f"Split sentence: {sentence.split()}")
```

Try these exercises to practice string manipulation:

```python
# Exercise 1: Create a function that takes a name and returns a greeting
def create_greeting(name):
    return f"Hello, {name.capitalize()}! Welcome to Python!"

# Test it
print(create_greeting("alice"))
```

```python
# Exercise 2: Count vowels in a string
def count_vowels(text):
    vowels = 'aeiouAEIOU'
    count = sum(1 for char in text if char in vowels)
    return count

# Test it
test_string = "Hello World"
print(f"Vowels in '{test_string}': {count_vowels(test_string)}")
```
