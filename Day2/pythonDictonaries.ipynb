# Python Dictionaries Tutorial

A Dictionary is a collection which is unordered, changeable and indexed. No duplicate members.

## Creating Dictionaries

```python
# Create dict
person = {
    'first_name': 'John',
    'last_name': 'Doe',
    'age': 30
}

print("Created person dictionary:")
print(person)

# Use constructor (alternative method)
# person2 = dict(first_name='Sara', last_name='Williams')
# print("Dictionary using constructor:", person2)
```

## Accessing Dictionary Values

```python
# Get value using square brackets
print("First name (using []):", person['first_name'])

# Get value using .get() method (safer - returns None if key doesn't exist)
print("Last name (using .get()):", person.get('last_name'))

# Demonstrate safety of .get() method
print("Non-existent key with []:", end=" ")
try:
    print(person['middle_name'])
except KeyError:
    print("KeyError raised!")

print("Non-existent key with .get():", person.get('middle_name'))
```

## Modifying Dictionaries

```python
# Add key/value
person['phone'] = '555-555-5555'
print("After adding phone:", person)

# Update existing value
person['age'] = 31
print("After updating age:", person)
```

## Dictionary Keys, Values, and Items

```python
# Get dict keys
print("Dictionary keys:", person.keys())

# Get dict values
print("Dictionary values:", person.values())

# Get dict items (key-value pairs)
print("Dictionary items:", person.items())
```

## Copying Dictionaries

```python
# Copy dict
person2 = person.copy()
person2['city'] = 'Boston'

print("Original person:", person)
print("Copied person2:", person2)
print("Notice: Changes to person2 don't affect person")
```

## Removing Items from Dictionaries

```python
print("Before removal:", person)

# Remove item using del
del(person['age'])
print("After del(person['age']):", person)

# Remove item using pop (returns the removed value)
removed_phone = person.pop('phone')
print(f"After pop('phone'): {person}")
print(f"Removed value: {removed_phone}")
```

## Dictionary Length and Clearing

```python
# Get length
print("Length of person2:", len(person2))
print("person2 contents:", person2)

# Clear dictionary
person.clear()
print("After clearing person:", person)
print("Length of cleared person:", len(person))
```

## List of Dictionaries

Dictionaries are often used in lists to represent collections of structured data.

```python
# List of dict
people = [
    {'name': 'Martha', 'age': 30},
    {'name': 'Kevin', 'age': 25}
]

print("List of people:", people)

# Access specific person's name
print("Second person's name:", people[1]['name'])

# Loop through list of dictionaries
print("\nAll people:")
for i, person in enumerate(people):
    print(f"Person {i+1}: {person['name']}, age {person['age']}")
```

## Advanced Dictionary Operations

```python
# Dictionary comprehension
numbers = {x: x**2 for x in range(1, 6)}
print("Dictionary comprehension (squares):", numbers)

# Merge dictionaries (Python 3.9+)
dict1 = {'a': 1, 'b': 2}
dict2 = {'c': 3, 'd': 4}
merged = dict1 | dict2  # Or use {**dict1, **dict2} for older Python versions
print("Merged dictionaries:", merged)

# Check if key exists
print("'name' in people[0]:", 'name' in people[0])
print("'salary' in people[0]:", 'salary' in people[0])
```

## Summary

Key dictionary concepts covered:

1. **Creation**: Using `{}` syntax or `dict()` constructor
2. **Access**: Using `[]` or `.get()` method
3. **Modification**: Adding, updating, and removing key-value pairs
4. **Methods**: `.keys()`, `.values()`, `.items()`, `.copy()`, `.pop()`, `.clear()`
5. **Complex structures**: Lists of dictionaries for structured data
6. **Safety**: `.get()` method prevents KeyError exceptions
