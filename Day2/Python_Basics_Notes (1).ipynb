# Introduction to Python

Python is a high-level, interpreted programming language known for its simplicity and readability. It's great for beginners and powerful enough for professionals.


# Variables and Data Types

Variables store data. In Python, you don't need to declare the type.

```python
name = "Alice"       # string
age = 25             # integer
height = 5.6         # float
is_student = True    # boolean
```

##  Common Data Types
- `str` Text data
- `int` Whole numbers
- `float`  Decimal numbers
- `bool`  True/False values

# Strings

Strings are text inside quotes.

```python
greeting = "Hello, world!"
```

## Useful Methods
```python
print(greeting.upper())     # HELLO, WORLD!
print(greeting.lower())     # hello, world!
print(greeting.replace("world", "Python"))  # Hello, Python!
```

# Numbers and Math

Python can do all basic math operations.

```python
a = 10
b = 3

print(a + b)  # 13
print(a - b)  # 7
print(a * b)  # 30
print(a / b)  # 3.333...
print(a // b) # 3
print(a % b)  # 1
print(a ** b) # 1000
```

# Lists

A list holds multiple items.

```python
fruits = ["apple", "banana", "cherry"]
print(fruits[0])  # apple
```

## Modify Lists
```python
fruits.append("orange")
fruits.remove("banana")
print(fruits)
```
