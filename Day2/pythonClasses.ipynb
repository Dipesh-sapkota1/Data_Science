# Python Classes and Object-Oriented Programming

A class is like a blueprint for creating objects. An object has properties and methods (functions) associated with it. Almost everything in Python is an object.

## Creating a Base Class

```python
# Create class
class User:
    # Constructor
    def __init__(self, name, email, age):
        self.name = name
        self.email = email
        self.age = age
        
        # Adding Encapsulation of variables... 
        # Encapsulation is the concept of making the variables non-accessible 
        # or accessible upto some extent from the child classes
        self._private = 1000  # Encapsulated variables are declared with '_' in the constructor.

    def greeting(self):
        return f'My name is {self.name} and I am {self.age}'

    def has_birthday(self):
        self.age += 1
 
    # Function for encap variable
    def print_encap(self):
        print(self._private)

print("User class created successfully!")
```

## Class Inheritance

We can extend classes to create child classes that inherit properties and methods from the parent class.

```python
# Extend class
class Customer(User):
    # Constructor
    def __init__(self, name, email, age):
        # Called proper parent class constructor to make this as proper child inheriting all methods
        User.__init__(self, name, email, age)
        self.name = name
        self.email = email
        self.age = age
        self.balance = 0

    def set_balance(self, balance):
        self.balance = balance

    # Method overriding - this will override the parent's greeting method
    def greeting(self):
        return f'My name is {self.name} and I am {self.age} and my balance is {self.balance}'

print("Customer class created successfully!")
```

## Creating and Using Objects

```python
# Init user object
brad = User('Brad Traversy', 'brad@gmail.com', 37)
print(f"Created user: {brad.name}, {brad.email}, {brad.age}")

# Init customer object
janet = Customer('Janet Johnson', 'janet@yahoo.com', 25)
print(f"Created customer: {janet.name}, {janet.email}, {janet.age}")
```

## Using Object Methods

```python
# Set balance for customer
janet.set_balance(500)
print("Janet's greeting:", janet.greeting())

# Call birthday method on user
brad.has_birthday()
print("Brad's greeting after birthday:", brad.greeting())
```

## Encapsulation Demonstration

Encapsulation allows us to hide internal data and control access to it. Variables prefixed with `_` are conventionally treated as private.

```python
# Encapsulation demonstration
print("\n=== Encapsulation Demo ===")
print("Brad's initial private variable:")
brad.print_encap()

# Changing for brad
brad._private = 800
print("Brad's private variable after change:")
brad.print_encap()

# Method inherited from parent
print("Janet's private variable (unchanged by Brad's modification):")
janet.print_encap()  # Changing the variable for brad doesn't affect janet's variable --> Encapsulation

janet._private = 600
print("Janet's private variable after change:")
janet.print_encap()

# Similarly changing janet's doesn't affect brad's variable
print("Brad's private variable (unchanged by Janet's modification):")
brad.print_encap()
```

## Summary

This notebook demonstrates key OOP concepts in Python:

1. **Class Creation**: Defining blueprints for objects
2. **Constructors**: Using `__init__` to initialize object properties
3. **Inheritance**: Creating child classes that extend parent classes
4. **Method Overriding**: Child classes can override parent methods
5. **Encapsulation**: Using `_` prefix to indicate private variables
6. **Object Independence**: Each object maintains its own state

```python
# Final state of our objects
print("=== Final Object States ===")
print(f"Brad: {brad.greeting()}, Private: {brad._private}")
print(f"Janet: {janet.greeting()}, Private: {janet._private}")
```
