# Python Variables Tutorial

## Introduction to Variables

Variables are containers for storing data values. In Python, variables are created when you assign a value to them:

```python
x = 10     # x is now a variable holding the integer value 10
print(x)
```

Unlike other programming languages, Python has no command for declaring a variable. A variable is created the moment you first assign a value to it.

## Variable Rules

```python
'''
VARIABLE RULES:
  - Variable names are case sensitive (name and NAME are different variables)
  - Must start with a letter or an underscore
  - Can have numbers but can not start with one
'''

# Valid variable names
name = "John"
_name = "John"
name_1 = "John"
NAME = "Different from name"

# Invalid variable names (uncomment to see errors)
# 1name = "John"  # Cannot start with a number
# my-name = "John"  # Cannot use hyphen
# my name = "John"  # Cannot use space
```

## Variable Data Types

Python has several core data types:

| Type | Description | Example |
|------|-------------|--------|
| int | Integer numbers | `x = 1` |
| float | Floating point numbers | `y = 2.5` |
| str | String (text) | `name = 'John'` |
| bool | Boolean (True/False) | `is_cool = True` |

```python
# Basic variable types
x = 1           # int
y = 2.5         # float
name = 'John'   # str
is_cool = True  # bool

# Print the variables and their types
print(f"x = {x}, type: {type(x)}")
print(f"y = {y}, type: {type(y)}")
print(f"name = {name}, type: {type(name)}")
print(f"is_cool = {is_cool}, type: {type(is_cool)}")
```

## Multiple Assignment

Python allows you to assign values to multiple variables in one line:

```python
# Multiple assignment
x, y, name, is_cool = (1, 2.5, 'John', True)

# Print to verify
print(x, y, name, is_cool)
```

## Basic Math Operations

You can perform basic math operations with variables:

```python
a = x + y   # Addition
b = x - y   # Subtraction 
c = x * y   # Multiplication
d = x / y   # Division

print(f"a = x + y = {a}")
print(f"b = x - y = {b}")
print(f"c = x * y = {c}")
print(f"d = x / y = {d}")

# Other common operations
e = x ** 2  # Exponentiation
f = x % 2   # Modulus (remainder)
g = x // y  # Floor division

print(f"e = x ** 2 = {e}")
print(f"f = x % 2 = {f}")
print(f"g = x // y = {g}")
```

## Type Conversion (Casting)

You can convert variables from one type to another with these functions:

```python
# Original values
x = 1
y = 2.5

# Type conversion
x_str = str(x)  # Convert to string
y_int = int(y)  # Convert to integer
z_float = float(x)  # Convert to float

# Print the results
print(f"x_str = {x_str}, type: {type(x_str)}")
print(f"y_int = {y_int}, type: {type(y_int)}")
print(f"z_float = {z_float}, type: {type(z_float)}")
```

## Bonus: Advanced Variable Concepts

Python variables actually reference objects. Understanding this will help you write better Python code.

```python
# Variable references
a = [1, 2, 3]  # a references a list object
b = a          # b now references the same list object

print(f"Original: a = {a}, b = {b}")

# Modifying the list through b also affects a
b.append(4)
print(f"After modification: a = {a}, b = {b}")

# Creating a copy (new reference)
c = a.copy()
c.append(5)
print(f"After copy modification: a = {a}, c = {c}")
```
