# Python Conditionals Tutorial

This notebook covers if/else conditions, comparison operators, logical operators, and membership operators in Python.

## Introduction to Conditionals

If/Else conditions are used to decide to do something based on something being true or false.

```python
# Set up variables for examples
x = 21
y = 20

print(f"x = {x}")
print(f"y = {y}")
```

## Comparison Operators

Comparison operators (==, !=, >, <, >=, <=) are used to compare values.

```python
# Demonstrate comparison operators
print(f"x == y: {x == y}")
print(f"x != y: {x != y}")
print(f"x > y: {x > y}")
print(f"x < y: {x < y}")
print(f"x >= y: {x >= y}")
print(f"x <= y: {x <= y}")
```

## Simple If Statement

```python
# Simple if
if x > y:
    print(f'{x} is greater than {y}')
```

## If/Else Statement

```python
# If/else
if x > y:
    print(f'{x} is greater than {y}')
else:
    print(f'{y} is greater than {x}')
```

## If/Elif/Else Statement

```python
# elif (else if)
if x > y:
    print(f'{x} is greater than {y}')
elif x == y:
    print(f'{x} is equal to {y}')
else:
    print(f'{y} is greater than {x}')
```

## Testing with Equal Values

```python
# Test with equal values
a = 10
b = 10

if a > b:
    print(f'{a} is greater than {b}')
elif a == b:
    print(f'{a} is equal to {b}')
else:
    print(f'{b} is greater than {a}')
```

## Nested If Statements

```python
# Nested if
if x > 2:
    if x <= 10:
        print(f'{x} is greater than 2 and less than or equal to 10')
    else:
        print(f'{x} is greater than 10')
else:
    print(f'{x} is less than or equal to 2')
```

## Logical Operators

Logical operators (and, or, not) are used to combine conditional statements.

### AND Operator

```python
# and operator - both conditions must be true
if x > 2 and x <= 10:
    print(f'{x} is greater than 2 and less than or equal to 10')
else:
    print(f'{x} does not meet both conditions')
```

### OR Operator

```python
# or operator - at least one condition must be true
if x > 25 or x < 30:
    print(f'{x} is either greater than 25 or less than 30 (or both)')
else:
    print(f'{x} is not greater than 25 and not less than 30')
```

### NOT Operator

```python
# not operator - negates the condition
if not(x == y):
    print(f'{x} is not equal to {y}')
else:
    print(f'{x} is equal to {y}')
```

## Membership Operators

Membership operators (in, not in) are used to test if a sequence is present in an object.

```python
numbers = [1, 2, 3, 4, 5]
print(f"Numbers list: {numbers}")
print(f"Testing with x = {x}")
```

### IN Operator

```python
# in operator
if x in numbers:
    print(f'{x} is in the numbers list')
else:
    print(f'{x} is not in the numbers list')

# Test with a number that is in the list
test_num = 3
if test_num in numbers:
    print(f'{test_num} is in the numbers list')
else:
    print(f'{test_num} is not in the numbers list')
```

### NOT IN Operator

```python
# not in operator
if x not in numbers:
    print(f'{x} is not in the numbers list')
else:
    print(f'{x} is in the numbers list')
```

## Identity Operators

Identity operators (is, is not) compare the objects, not if they are equal, but if they are actually the same object, with the same memory location.

### IS Operator

```python
# is operator
if x is y:
    print(f'{x} is the same object as {y}')
else:
    print(f'{x} is not the same object as {y}')

# Example with same values but different objects
list1 = [1, 2, 3]
list2 = [1, 2, 3]
list3 = list1

print(f"list1 == list2: {list1 == list2}")  # Same values
print(f"list1 is list2: {list1 is list2}")  # Different objects
print(f"list1 is list3: {list1 is list3}")  # Same object
```

### IS NOT Operator

```python
# is not operator
if x is not y:
    print(f'{x} is not the same object as {y}')
else:
    print(f'{x} is the same object as {y}')
```

## Practical Examples

Let's look at some real-world examples of conditionals.

```python
# Grade calculator
def get_letter_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

# Test the function
test_scores = [95, 87, 73, 65, 45]
for score in test_scores:
    grade = get_letter_grade(score)
    print(f"Score: {score} -> Grade: {grade}")
```

```python
# Age category classifier
def classify_age(age):
    if age < 0:
        return "Invalid age"
    elif age < 13:
        return "Child"
    elif age < 20:
        return "Teenager"
    elif age < 60:
        return "Adult"
    else:
        return "Senior"

# Test the function
test_ages = [5, 16, 25, 45, 70]
for age in test_ages:
    category = classify_age(age)
    print(f"Age: {age} -> Category: {category}")
```

```python
# Password strength checker
def check_password_strength(password):
    if len(password) < 6:
        return "Weak: Too short"
    elif len(password) < 8:
        return "Medium: Could be longer"
    elif any(char.isdigit() for char in password) and any(char.isupper() for char in password):
        return "Strong: Good combination"
    else:
        return "Medium: Add numbers and uppercase letters"

# Test the function
test_passwords = ["123", "password", "Password123", "mypassword"]
for pwd in test_passwords:
    strength = check_password_strength(pwd)
    print(f"Password: '{pwd}' -> {strength}")
```

## Practice Exercises

Try these exercises to practice working with conditionals:

```python
# Exercise 1: Number comparison
def compare_numbers(a, b, c):
    if a > b and a > c:
        return f"{a} is the largest"
    elif b > a and b > c:
        return f"{b} is the largest"
    elif c > a and c > b:
        return f"{c} is the largest"
    else:
        return "Some numbers are equal"

# Test it
print(compare_numbers(10, 20, 15))
print(compare_numbers(30, 10, 25))
print(compare_numbers(5, 5, 5))
```

```python
# Exercise 2: Season detector
def get_season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    elif month in [9, 10, 11]:
        return "Fall"
    else:
        return "Invalid month"

# Test it
for month in [1, 4, 7, 10, 13]:
    season = get_season(month)
    print(f"Month {month}: {season}")
```

```python
# Exercise 3: Login validator
def validate_login(username, password, users_db):
    if username not in users_db:
        return "Username not found"
    elif users_db[username] != password:
        return "Incorrect password"
    else:
        return "Login successful"

# Test database
users = {
    "alice": "password123",
    "bob": "secret456",
    "charlie": "mypass789"
}

# Test it
test_cases = [
    ("alice", "password123"),
    ("bob", "wrongpass"),
    ("david", "anypass")
]

for username, password in test_cases:
    result = validate_login(username, password, users)
    print(f"Login attempt - User: {username}, Result: {result}")
```
